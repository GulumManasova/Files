```python
# Add the two matrices element-wise
import numpy as np
m1=np.array([[1,4],[-5,8]])
m2=np.array([[7,-6],[-11,2]])
msum=m1+m2
print(msum)
```

```python
import numpy as np

n = 2
a = np.random.randn(n, n)
b = np.random.randn(n, n)
c = np.random.randn(n, n)

# Compute s1 and s2
s1 = np.dot(a, np.dot(b, c))  # Matrix multiplication: a * (b * c)
s2 = np.dot(np.dot(a, b), c)  # Matrix multiplication: (a * b) * c

# Print results
print("a:\n", a)
print("b:\n", b)
print("c:\n", c)
print("s1:\n", s1)
print("s2:\n", s2)

# Check if they are the same
are_equal = np.allclose(s1, s2)
print("s1 and s2 are equal:", are_equal)
```
