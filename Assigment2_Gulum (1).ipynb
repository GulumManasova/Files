## What is meant by data visualization?
### Data visualization is the representation of the data through use of common graphics such as charts,plots,infographics and even animations

## Why data visualization is important?
### The importance is analyzing complex data,identifying patterns ,and extracting valuable insights. Simplifying complex information and presenting it visually enables decision_makers to make informed and effective decision quickly and accurately

## Why do modern businesses need data visualization?
### Modern businesses need data visualization because it helps them make sense of complex data, identify trends, and make informed decisions quickly, Better Decision-Making,Enhanced Data Comprehension,Faster Insights,Improved Communication,Real-Time Monitoring,Competitive Advantage,Enhanced Storytelling.

## How is data visualization used for business decision making?

```python
import matplotlib.pyplot as plt
import pandas as pd


name = ['Ahmed', 'Omar', 'Ali', 'Zlad', 'Salwa', 'Ula']
mobile_sales = [2540, 1370, 1320, 2000, 2100, 2150]
tv_sales = [2200, 1900, 2150, 1850, 1770, 2000]


df = pd.DataFrame({'name': name, 'Mobile_sales': mobile_sales, 'TV_sales': tv_sales})


df.set_index('name')[['Mobile_sales', 'TV_sales']].plot(kind='bar', figsize=(8, 5))


plt.xlabel('Name')
plt.ylabel('Sales volume')
plt.title('Bar plot of sales volume')
plt.xticks(rotation=45)
plt.show()
```

```python
fig=plt.figure(figsize=(10,6))
plt.pie(mobile_sales,labels=name,autopct="%1.1f%%")
plt.title('Mobile_sales')
plt.show()
```

```python
fig=plt.figure(figsize=(12,6))
plt.pie(tv_sales,labels=name,autopct='%1.1f%%')
plt.title('TV_sales')
plt.show()
```

```python

plt.boxplot([mobile_sales, tv_sales], labels=['Mobile sales', 'TV sales'], patch_artist=True)


plt.ylabel('item sales')
plt.title('Box plot of item sales')
plt.show()
```

```python

plt.fill_between(name, mobile_sales, color='black', alpha=0.9,label='Mobile_sales')
plt.fill_between(name, tv_sales, color='grey', alpha=0.9,label='TV_sales')


plt.xlabel('name')
plt.ylabel('item sales')
plt.title('Area plot')
plt.legend()
plt.show()
```

```python
# Create stacked bar plot
df.set_index('name')[['mobile_sales', 'tv_sales']].plot(kind='bar', stacked=True, figsize=(8, 5), color=['orange', 'black'])

# Add labels and title
plt.xlabel('Name')
plt.ylabel('Item Sales')
plt.title('Stacked Bar plot')
plt.xticks(rotation=45)
plt.show()
```
